```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 10000)
```

```python
def my_fxn(my_x):
    my_y = (my_x**2 - 1)/(my_x - 1)
    return my_y
```

```python
my_fxn(0)
```

```python
y = my_fxn(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.axvline(x=0, color='purple', linestyle='--')
plt.axhline(y=2, color='purple', linestyle='--')
_ = ax.plot(x,y)
```

```python
def my_fxn(my_x):
    my_y = (my_x**2 - 25)/(my_x + 1)
    return my_y
```

```python
y = my_fxn(x)
```

```python
my_fxn(5)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(1, 8)
plt.ylim(-2, 3)
plt.axvline(x=5, color='purple', linestyle='--')
plt.axhline(y=0, color='purple', linestyle='--')
_ = ax.plot(x,y)
```

```python
def my_fxn(my_x):
    my_y = (my_x**2 - 2*my_x -8)/(my_x - 4)
    return my_y
```

```python
y = my_fxn(x)
```

```python
# my_fxn(4) ZeroDivisionError: division by zero
```

```python
def inf_fxn(my_x):
    my_y = 25/my_x
    return my_y
```

```python
inf_fxn(1e3)
```

```python
inf_fxn(1e6)
```

```python
y = inf_fxn(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
_ = ax.plot(x, y)
```
