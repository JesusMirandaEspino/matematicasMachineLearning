```python

def f(x):
  return x**2


x_0 = 2


x_values = np.array([1.9, 1.99, 1.999, 2.001, 2.01, 2.1])


y_values = f(x_values)


for x, y in zip(x_values, y_values):
  print(f"f({x}) = {y}")


limite_esperado = 4
diferencias = np.abs(y_values - limite_esperado)


for x, diff in zip(x_values, diferencias):
    print(f"|f({x}) - {limite_esperado}| = {diff}")
```

```python
x = np.linspace(-5, 5, 100)

```

```python
h = 0.001  # Tamaño del paso
df_dx = (f(x + h) - f(x - h)) / (2 * h)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label="f(x) = x^2")
plt.plot(x, df_dx, label="Derivada (aproximación)")
plt.xlabel("x")
plt.ylabel("f(x) / df_dx")
plt.title("Función y su Derivada")
plt.legend()
plt.grid(True);
```
