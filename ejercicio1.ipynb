```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000)
```

```python
t
```

```python
d_r = 2.5 * t
```

```python
d_r
```

```python
d_s = 3 * (t-5)
```

```python
d_s
```

```python
fig, ax = plt.subplots()
plt.title('El Ladrón de Bancos Atrapado')
plt.xlabel('tiempo (en minutos)')
plt.ylabel('distancia (en km)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```

```python
t_kj = np.linspace(0, 80, 1000)
```

```python
m_1 = t_kj
```

```python
m_2 = []
day_value = 0
for day in  t_kj:
    #day_value = 4( day - 30 )
    inicial = day - 30
    day_value = inicial * 4
    m_2.append(day_value)
```

```python
fig, ax = plt.subplots()
plt.title('mark1 y mark 2')
plt.xlabel('tiempo (en dias)')
plt.ylabel('energia (en kj)')
ax.set_xlim([0, 80])
ax.set_ylim([0, 200])
ax.plot(t_kj, m_1, c='green')
ax.plot(t_kj, m_2, c='brown');
```
